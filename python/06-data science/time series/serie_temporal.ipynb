```python
import pandas as pd
from pandas.plotting import autocorrelation_plot
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plotar(titulo, labelx, labely, x, y, df):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x, y=y, data=df)
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc='left', fontsize=18)
    ax.set_xlabel(labelx, fontsize=14)
    ax.set_ylabel(labely, fontsize=14)
    ax = ax

def plot_comparacao(x, y1, y2, y3, df, titulo):
    plt.figure(figsize=(16,12))
    ax = plt.subplot(3,1,1)
    ax.set_title(titulo, fontsize=18, loc='left')
    sns.lineplot(x=x, y=y1, data=df)
    plt.subplot(3,1,2)
    sns.lineplot(x=x, y=y2, data=df)
    plt.subplot(3,1,3)
    sns.lineplot(x=x, y=y3, data=df)
    ax = ax

def auto_correlacao(titulo, df):
    ax = plt.figure(figsize=(12,4.5))
    ax.suptitle(titulo, fontsize=18, x=0.26, y=0.95)
    autocorrelation_plot(df)
    ax = ax
```

```python
alucar = pd.read_csv('alucar.csv')
print(f'O arquivo "alucar.csv" possui {alucar.shape[0]} registros, com {alucar.shape[1]} colunas de variáveis\n'
      f'Quantidade de dados nulos: {alucar.isna().sum().sum()}')
alucar.sample(5)
```

```python
alucar['mes'] = pd.to_datetime(alucar['mes'])
alucar.dtypes
```

```python
alucar['aumento'] = alucar['vendas'].diff()
alucar['aceleracao'] = alucar['aumento'].diff()
print(f'Criando novas variáveis com os valores de aumento e aceleração de aumento das vendas:')
alucar.head(5)
```

```python
plotar(titulo='Vendas da Alucar de 2017 e 2018', labelx='Tempo', labely='Vendas (R$)', x='mes', y='vendas', df=alucar)
```

```python
plotar(titulo='Aumento das vendas da Alucar de 2017 e 2018', labelx='Tempo', labely='Aumento', x='mes', y='aumento', df=alucar)
```

```python
plotar(titulo='Aceleração no aumento das vendas da Alucar de 2017 e 2018', labelx='Tempo', labely='Aceleração', x='mes', y='aceleracao', df=alucar)
```

```python
plot_comparacao(x='mes', y1='vendas', y2='aumento', y3='aceleracao', df=alucar, titulo='Análise das vendas da Alucar de 2017 e 2018')
```

```python
auto_correlacao(titulo='Correlação das vendas', df=alucar['vendas'])
auto_correlacao(titulo='Correlação do aumento', df=alucar['aumento'][1:])
auto_correlacao(titulo='Correlação da aceleração', df=alucar['aceleracao'][2:])
```

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')[['vendas', 'aumento', 'aceleracao']].mean().round()
vendas_agrupadas
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
resultado = seasonal_decompose(chocolura['vendas'], period=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
data = ({'observacao':observacao, 'tendencia':tendencia, 'sazonalidade':sazonalidade, 'ruido':ruido})

resultado = pd.DataFrame(data)
resultado.head()
```

```python
plot_comparacao(x=resultado.index, y1='observacao', y2='tendencia', y3='sazonalidade', df=resultado, titulo='Exemplo de Statsmodels')
```
